```python
import os
import tabula
import numpy as np
import pandas as pd
```

```python
def listdir_ignore_hidden(dirname: str) -> list:
    """
    Function that gets the name of all non-hidden filenames in a directory
    
    @param dirname: name of directory to pull the filenames from
    @return: non-
    """
    return [f for f in os.listdir(dirname) if not f.startswith('.')]
```

```python
def handleDF(dfList: list) -> pd.DataFrame:
    """
    returns a single column dataframe from a varying column-length list of dataframes
    
    @param dfList: a list of dataframes
    @return: single column dataframe
    """
    singleColumn = []
    for df in dfList:
        if len(df.columns) == 0:
            pass
        elif len(df.columns) == 1:
            singleColumn.append(df)
        elif len(df.columns) == 2:
            tempDF = (df.iloc[:,0] + " " + df.replace(np.NaN, "").iloc[:, 1]).str.strip()
            singleColumn.append(tempDF)
        elif len(df.columns) == 3:
            tempDF = (df.iloc[:,0] + " " + df.replace(np.NaN, "").iloc[:, 1] + " " + df.replace(np.NaN, "").iloc[:, 2]).str.strip()
            singleColumn.append(tempDF)
        elif len(df.columns) == 4:
            tempDF = (df.iloc[:,0] + " " + df.replace(np.NaN, "").iloc[:, 1] + " " + df.replace(np.NaN, "").iloc[:, 2] + " " + df.replace(np.NaN, "").iloc[:, 3]).str.strip()
            singleColumn.append(tempDF)
        elif len(df.columns) == 5:
            tempDF = (df.iloc[:,0] + " " + df.replace(np.NaN, "").iloc[:, 1] + " " + df.replace(np.NaN, "").iloc[:, 2] + " " + df.replace(np.NaN, "").iloc[:, 3] + " " + df.replace(np.NaN, "").iloc[:, 4]).str.strip()
            singleColumn.append(tempDF)
        elif len(df.columns) == 6:
            tempDF = (df.iloc[:,0] + " " + df.replace(np.NaN, "").iloc[:, 1] + " " + df.replace(np.NaN, "").iloc[:, 2] + " " + df.replace(np.NaN, "").iloc[:, 3] + " " + df.replace(np.NaN, "").iloc[:, 4] + " " + df.replace(np.NaN, "").iloc[:, 5]).str.strip()
            singleColumn.append(tempDF)
        else:
            raise Exception("unexpected column length")
    return pd.concat(singleColumn)
```

```python
# File Extraction
initDir = "CrimeLogPDF"
yearDirs = [os.path.join(initDir, yearDir) for yearDir in listdir_ignore_hidden(initDir)]
monthDirs = [os.path.join(yearDir, monthDir) for yearDir in yearDirs for monthDir in listdir_ignore_hidden(yearDir)]

# type1 has different format compared to type2
type1Dirs = ["CrimeLogPDF/2011/JANUARY 2011", "CrimeLogPDF/2011/FEBRUARY 2011", "CrimeLogPDF/2011/MARCH 2011", "CrimeLogPDF/2011/APRIL 2011", "CrimeLogPDF/2011/MAY 2011", "CrimeLogPDF/2011/JUNE 2011"]
type2Dirs = [monthDir for monthDir in monthDirs if monthDir not in type1Dirs]

type1Files = [os.path.join(type1Dir, file) for type1Dir in type1Dirs for file in listdir_ignore_hidden(type1Dir)]
type2Files = [os.path.join(type2Dir, file) for type2Dir in type2Dirs for file in listdir_ignore_hidden(type2Dir)]
type2Files.remove('CrimeLogPDF/2012/December 2012/Dec 1, 2012.pdf') # A corrupted pdf file that needs to be removed
```

```python
# conversion to single column dataframe
i = 0
area = (65.27,9.2,791.47,560.6)
for file in type2Files:
    i += 1
    print(f"{i}/{len(type2Files)}")
    df = handleDF(tabula.read_pdf(file, pages="all", area=area, multiple_tables=True))
    df.to_csv('singleColumn.csv', mode='a', index=False, header=False)
```
